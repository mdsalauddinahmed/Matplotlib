```python
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "java"]
y = [85, 70, 68, 82]
plt.xlabel("language", fontsize=20)
plt.ylabel("NO.")
plt.title("SalauddinAmir",fontsize=20)
c=["r","y","b","g"]

plt.bar(x, y,width=.4,color=c,align="edge")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "java"]
y = [85, 70, 68, 82]
z=[20,30,40,50]
plt.xlabel("language", fontsize=20)
plt.ylabel("NO.")
plt.title("SalauddinAmir",fontsize=20)


plt.bar(x, y,width=.4,color="r",edgecolor="r",label="Amir")
plt.bar(x, z,width=.4,color="y",label="Salauddin")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = ["python", "c", "c++", "java"]
y = [85, 70, 68, 82]
z=[20,30,40,50]

width=.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language", fontsize=20)
plt.ylabel("NO.")
plt.title("SalauddinAmir",fontsize=20)


plt.bar(p, y,width,color="r",label="Amir")
plt.bar(p1, z,width,color="y",label="Salauddin")
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = ["python", "c", "c++", "java"]
y = [85, 70, 68, 82]
z=[20,30,40,50]

width=.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language", fontsize=20)
plt.ylabel("NO.")
plt.title("SalauddinAmir",fontsize=20)


plt.barh(p, y,width,color="r",label="Amir")
plt.barh(p1, z,width,color="y",label="Salauddin")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
